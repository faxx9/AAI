```python
#  take a mean of the following columns


columns=["Age","Cholesterol","Income"]

for i in columns:
    # first we find number of a value
    num=len(df[i])
    #sum of a columns
    sum=df[i].sum()
    mean=sum/num
    #then use print funtion
    print(f"mean of {i} : {mean}")
```

```python
# Select the columns that we need a standard deviation
columns=["Age","Cholesterol","Income"]
#using for loop
for i in columns:
    #find length of data
    n=len(df[i])
    #values assign to veriable x
    x=df[i]
    #mean of a each columns by numpy
    x_mean=np.mean(df[i])
    #subtract mean value of columns from origional value
    subt=x-x_mean
    #squre of a data
    minus_square=subt**2
    # After that we apply summation
    sum_of_square=np.sum(minus_square)
    division=sum_of_square/n
    #take SD
    SD=np.sqrt(division)
    print(f"\n\n Standard deviation of {i}: {SD}")

```

```python
nunique_10=df.nunique()>10
```

```python
# Get the columns with uniqye values more than 10
cols_to_scale=df.loc[:, nunique_10].columns
print(cols_to_scale)
```

```python
# Chect the average of value counts in imbalce i data
y.value_counts(normalize=True)*100
```

```python
# use the Smoth in which classes has equel representation

from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=365)
x_smote,y_smote=smote.fit_resample(x,y)
```

```python
#Chect the values counts of y smote

y_smote.value_counts(normalize=True)
```

```python
#visuize the values of y mote with pie plot

plt.pie(y_smote.value_counts(), autopct="%1.1f%%", startangle=140)
plt.show()
```

```python
# Again split the data on y smote and x smote

x_train,x_test,y_train,y_test=train_test_split(x_smote,y_smote, test_size=0.2)
```

```python
# scale the vaues of cols usin Rabust Scaler

from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
scaler.fit(x_train[cols_to_scale])
x_train[cols_to_scale]=scaler.transform(x_train[cols_to_scale])
x_test[cols_to_scale]=scaler.transform(x_test[cols_to_scale])
```

```python
# Import the modules
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#train gussain Model
clf=GaussianNB()
clf.fit(x_train,y_train)
clf_score=clf.score(x_test,y_test)
print(clf_score)
```

```python
# Logistic Regression
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr_score=lr.score(x_test,y_test)
print(lr_score)
```

```python
#Decistion tree model
dtr=DecisionTreeClassifier()
dtr.fit(x_train,y_train)
dtr_score=dtr.score(x_test,y_test)
print(dtr_score)
```

```python
#randomForestClasssifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf_score=rf.score(x_test,y_test)
print(rf_score)
```

```python
#kNeighborsClassifier
kon=KNeighborsClassifier()
kon.fit(x_train,y_train)
kon_score=kon.score(x_test,y_test)
print(kon_score)
```

```python
# Get prediction from these model
#store the key valuies of models
models=[clf,lr,dtr,rf,kon]
model_prediction={}
for model in models:
    model_name=str(model).split("(")[0]
    model_prediction[model_name]=model.predict(x_test)
    print(model_prediction)
```

```python
# Evaluate the perforamnce model
from sklearn.metrics import accuracy_score,\
                                           f1_score, precision_score,recall_score

model_evaluation={}
for model, preds in model_prediction.items():
    model_evaluation[model]=[
        round(accuracy_score(y_test,preds)*100,2),
        round(f1_score(y_test,preds)*100,2),
        round(precision_score(y_test,preds)*100,2),
        round(recall_score(y_test,preds)*100,2),
    ]
```

```python
#Get the results
results_df=pd.DataFrame(model_evaluation, index=["Accuracy","F1 Score","Precision","recall"])
results_df.style.background_gradient(axis=None, cmap="bone")
```
